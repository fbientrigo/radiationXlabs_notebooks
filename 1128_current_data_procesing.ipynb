## Teoria:

- **‚ÄúOff‚Äù** (latch-up/protecci√≥n) ocurre cuando `IDC ‚âà 0 A`. De hecho, el percentil 5% y 1% son 0 A.  
- **‚ÄúIdle principal‚Äù** se ubica alrededor de 0.5 A (50% y 75% en ~1.257‚Äì1.260).  

Para detectar **ca√≠das repentinas de ‚Äúidle principal‚Äù ‚Üí ‚Äúoff‚Äù** definimos:

- Umbral de **off**:  
  $$ \text{OFF\_TH} = 0.01\;\text{A} $$
  (cualquier corriente menor a 0.01 A la consideramos apagado / latch-up).

- Rango de **idle principal**:  
  $$ \text{IDLE\_LOW} = p(50)\;\text{A},\quad \text{IDLE\_HIGH} = p(75)\;\text{A} $$
  (todo valor en [ p(50), p(75)] A lo consideramos ‚Äúen operaci√≥n normal‚Äù).

De este modo nos aseguramos de que s√≥lo se detecten transiciones cuando est√° realmente en modo ‚Äúidle principal‚Äù y cae a ‚Äúoff‚Äù, ignorando el estado intermedio (~0.65 A) que corresponde a otras fases de consumo.


```python
# Definici√≥n de umbrales basados en el an√°lisis distribuido de IDC

percentile = lambda p: np.percentile(df["IDC"].dropna(), p)

OFF_TH = percentile(p_off)       # < 0.01 A ‚Üí se considera ‚Äúoff‚Äù (protecci√≥n/latch-up)
IDLE_LOW = percentile(p_idle_low)     # ‚â• 0.45 A
IDLE_HIGH = percentile(p_idle_high)   # ‚â§ 0.62 A
```

```python


# Extraer timestamps de eventos ‚Äúidle principal ‚Üí off‚Äù
events = []
prev_idle = False

# Convertimos la serie ‚ÄúIDC‚Äù en un array para evitar m√∫ltiples indexaciones
idc_arr = df["IDC"].values
times  = df.index.to_numpy()

for i in range(1, len(idc_arr)):
    prev = idc_arr[i - 1]
    curr = idc_arr[i]
    t    = times[i]
    
    # Estado ‚Äúidle principal‚Äù en la muestra previa
    if IDLE_LOW <= prev <= IDLE_HIGH:
        prev_idle = True
    else:
        prev_idle = False
    
    # Detectar transici√≥n ‚Äúidle principal ‚Üí off‚Äù
    if prev_idle and (curr < OFF_TH):
        # Buscar recuperaci√≥n: siguiente instant√°nea con IDC ‚â• IDLE_LOW
        # (a partir de i+1 en adelante)
        resto_idx = np.where(idc_arr[i+1:] >= IDLE_LOW)[0]
        if resto_idx.size > 0:
            j = resto_idx[0] + (i + 1)
            t_rec = times[j]
            duration_s = (t_rec - t).astype('timedelta64[s]').astype(int)
            events.append({
                "Time_event": t,
                "Duraci√≥n_s": duration_s,
                "Idle_previo": prev,
                "Off_detectado": curr,
                "Recuperaci√≥n_en": t_rec,
                "IDC_recuperaci√≥n": idc_arr[j]
            })

# Validaciones y salida
if not events:
    print("--> No se detectaron transiciones ‚Äòidle principal ‚Üí off‚Äô.")
else:
    eventos_df = pd.DataFrame(events)
    print(f"Se detectaron {len(events)} eventos de latch-up.\n")
    print("Primeros 5 eventos:\n", eventos_df.head())


```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot_latchups_on_current(df: pd.DataFrame,
                             eventos_df: pd.DataFrame,
                             current_col: str = "IDC",
                             off_color: str = "red",
                             recovery_color: str = "green",
                             span_alpha: float = 0.2,
                             figsize: tuple = (12, 5)) -> None:
    """
    Grafica la serie temporal de corriente y marca visualmente los eventos de latch-up.

    Para cada evento en `eventos_df` (index como Time_event y columna 'Recuperaci√≥n_en'),
    dibuja:
      - Una franja sombreada desde Time_event hasta Recuperaci√≥n_en (alpha=span_alpha).
      - Un marcador rojo en el instante Time_event.
      - Un marcador verde en el instante Recuperaci√≥n_en.

    Par√°metros
    ----------
    df : pandas.DataFrame
        DataFrame indexado por tiempo (datetime64[ns]) que contiene la columna de corriente.
    eventos_df : pandas.DataFrame
        DataFrame de eventos de latch-up, indexado por 'Time_event' y con columna
        'Recuperaci√≥n_en'.
    current_col : str, opcional
        Nombre de la columna con la corriente (por defecto "IDC").
    off_color : str, opcional
        Color para los marcadores de "off" (por defecto "red").
    recovery_color : str, opcional
        Color para los marcadores de recuperaci√≥n (por defecto "green").
    span_alpha : float, opcional
        Transparencia de las franjas sombreadas que indican duraci√≥n de latch-up.
    figsize : tuple, opcional
        Tama√±o de la figura (en pulgadas).

    Devuelve
    -------
    None
        Muestra la figura con la serie de corriente y las anotaciones de latch-up.

    Ejemplo de uso
    --------------
    >>> plot_latchups_on_current(df, eventos_df, current_col="IDC")
    """

    # Verificar √≠ndices datetime
    if not pd.api.types.is_datetime64_any_dtype(df.index):
        raise ValueError("El √≠ndice de `df` debe ser datetime64[ns].")

    if not pd.api.types.is_datetime64_any_dtype(eventos_df.index):
        raise ValueError("El √≠ndice de `eventos_df` (Time_event) debe ser datetime64[ns].")

    if "Recuperaci√≥n_en" not in eventos_df.columns:
        raise KeyError("`eventos_df` debe tener columna 'Recuperaci√≥n_en'.")

    # Preparar la figura
    fig, ax = plt.subplots(figsize=figsize, constrained_layout=True)

    # Graficar curva de corriente
    ax.plot(df.index, df[current_col], color="tab:blue", linewidth=1, label=current_col)
    ax.set_xlabel("Time")
    ax.set_ylabel(f"{current_col} [A]")
    ax.set_title("Serie de corriente con eventos de latch-up")
    ax.grid(True, linestyle="--", alpha=0.4)

    # Para cada evento, sombrear la duraci√≥n y marcar puntos
    for t_event, row in eventos_df.iterrows():
        t_recov = row["Recuperaci√≥n_en"]

        # Sombra desde el evento hasta la recuperaci√≥n
        ax.axvspan(t_event, t_recov, color="lightgrey", alpha=span_alpha)

        # Marcador rojo en Time_event
        ax.scatter(t_event, row["Off_detectado"], color=off_color, s=30, zorder=5,
                   label="Latch-up" if "Latch-up" not in ax.get_legend_handles_labels()[1] else "")

        # Marcador verde en Recuperaci√≥n_en
        rec_y = row["IDC_recuperaci√≥n"]
        ax.scatter(t_recov, rec_y, color=recovery_color, s=30, marker="s", zorder=5,
                   label="Recuperaci√≥n" if "Recuperaci√≥n" not in ax.get_legend_handles_labels()[1] else "")

    # Formateo de fechas en eje X
    ax.xaxis.set_major_locator(plt.MaxNLocator(6))
    fig.autofmt_xdate(rotation=30)

    # Leyenda y mostrar
    ax.legend(loc="upper right")
    plt.show()


# ======= Eecuci√≥n con datos =======


plot_latchups_on_current(df, eventos_df, current_col="IDC")
```

# Modelado de Resiliencia por Latch-Up

Definimos la variable aleatoria \(D\) = ‚Äúdosis hasta latch-up‚Äù. A partir de los datos obtenidos cruzando DMM vs beam, extraemos el vector \(\{d_i\}\).

- **Exponencial (tasa constante):**  
  $$f(d) = \lambda e^{-\lambda d}, \quad S(d) = e^{-\lambda d}$$  
  $\displaystyle \hat\lambda = 1 / \bar d$.

- **Weibull (envejecimiento o efecto inicial):**  
  $$f(d) = \frac{k}{\eta}\Bigl(\frac{d}{\eta}\Bigr)^{k-1}e^{-(d/\eta)^k}, \quad S(d) = e^{-(d/\eta)^k}$$  
  $\;k\neq 1$ ajusta ‚Äúdrift‚Äù de la tasa de falla.

El pipeline espera un arreglo NumPy con dosis al primer latch-up (Gy). Si el formato es distinto (por ejemplo, strings en lugar de floats, o timestamps mal convertidos), las funciones de ajuste fallar√°n. En la siguiente celda definimos las funciones y en la √∫ltima hacemos comprobaciones sobre el formato de `dose_to_failure` antes de ajustar los modelos.


```python
# --- Funciones de Ajuste y Supervivencia ---

import numpy as np
import scipy.stats as st

def fit_exponential(doses, alpha=0.05):
    """Ajusta exponencial: retorna ŒªÃÇ y su IC95%."""
    if len(doses)==0:
        raise ValueError("El arreglo de dosis est√° vac√≠o.")
    if np.any(doses <= 0):
        raise ValueError("Todas las dosis deben ser positivas.")
    lam_hat = 1.0 / doses.mean()
    n = len(doses)
    sum_d = doses.sum()
    chi2_l = st.chi2.ppf(alpha/2, 2*n)
    chi2_u = st.chi2.ppf(1 - alpha/2, 2*n)
    lam_ci = (chi2_l/(2*sum_d), chi2_u/(2*sum_d))
    return lam_hat, lam_ci

def fit_weibull(doses, n_boot=1000, alpha=0.05, seed=42):
    """Ajusta Weibull (shape k, scale Œ∑) y calcula IC95% por bootstrap."""
    if len(doses)==0:
        raise ValueError("El arreglo de dosis est√° vac√≠o.")
    if np.any(doses <= 0):
        raise ValueError("Todas las dosis deben ser positivas.")
    k_hat, loc_hat, eta_hat = st.weibull_min.fit(doses, floc=0)
    # Bootstrap para IC
    rng = np.random.default_rng(seed)
    n = len(doses)
    params = np.zeros((n_boot, 2))
    for i in range(n_boot):
        sample = rng.choice(doses, size=n, replace=True)
        k_b, loc_b, eta_b = st.weibull_min.fit(sample, floc=0)
        params[i] = (k_b, eta_b)
    k_ci = np.percentile(params[:,0], [100*alpha/2, 100*(1-alpha/2)])
    eta_ci = np.percentile(params[:,1], [100*alpha/2, 100*(1-alpha/2)])
    return (k_hat, eta_hat), (tuple(k_ci), tuple(eta_ci))

def S_exponential(d, lam):
    return np.exp(-lam * d)

def S_weibull(d, k, eta):
    return np.exp(-(d/eta)**k)


# ---  Ejecuci√≥n  ---
# dose_to_failure = np.array([...])  # Dosis en Gy
lam_hat, lam_ci = fit_exponential(dose_to_failure)
print("ŒªÃÇ =", lam_hat, "IC95% Œª:", lam_ci)

(k_hat, eta_hat), (k_ci, eta_ci) = fit_weibull(dose_to_failure)
print("kÃÇ =", k_hat, "Œ∑ÃÇ =", eta_hat, "IC95% k:", k_ci, "IC95% Œ∑:", eta_ci)

d_vals = np.array([0.5, 1.0, 2.0])
print("S_exp:", S_exponential(d_vals, lam_hat))
print("S_weibull:", S_weibull(d_vals, k_hat, eta_hat))
```
